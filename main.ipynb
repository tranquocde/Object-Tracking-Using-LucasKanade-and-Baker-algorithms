```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import RectBivariateSpline

# Generate some sample data
x = np.linspace(0, 4*np.pi, 10)
y = np.linspace(0, 2*np.pi, 5)
X, Y = np.meshgrid(x, y) #(5,10)

Z = np.sin(X) + np.cos(Y)# (5,10)
Z_ = Z.T
# # Define the grid for evaluation
x_new = np.linspace(0, 4*np.pi, 100)
y_new = np.linspace(0, 2*np.pi, 100)
X_new, Y_new = np.meshgrid(x_new, y_new)
print("X_new , Y_new shapes : ",X_new.shape,Y_new.shape)
# # Create a bivariate spline
spline = RectBivariateSpline(x, y, Z_)

# # Evaluate the spline on the new grid
Z_new = spline(x_new, y_new) #2D np array

# # Plot the original data
plt.figure(figsize=(8, 6))
plt.pcolormesh(X, Y, Z, cmap='viridis')
plt.colorbar(label='Z')
plt.title('Original Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# # Plot the interpolated data
plt.figure(figsize=(8, 6))
plt.pcolormesh(X_new, Y_new, Z_new, cmap='viridis')
plt.colorbar(label='Z')
plt.title('Interpolated Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
